## Calculating percentage of missing values for each column having null values

```python
percent_rate = df['rate'].isnull().sum()*100/df.shape[0]
percent_location = df['location'].isnull().sum()*100/df.shape[0]
percent_rest_type = df['rest_type'].isnull().sum()*100/df.shape[0]
percent_cuisines = df['cuisines'].isnull().sum()*100/df.shape[0]
percent_approx_cost = df['approx_cost(for two people)'].isnull().sum()*100/df.shape[0]
print(f'Rate percentage: {percent_rate}\nLocation percentage: {percent_location}\nRest_type percentage: {percent_rest_type}\nCuisines percentage: {percent_cuisines}\nApprox cost for two people percentage: {percent_approx_cost}')
```

## Treating missing values

```python
np.where(df['rate'] == 'NEW')
```

```python
df['rate'].replace('NEW',value = 0.0,inplace = True)
```

```python
np.where(df['rate'] == 'NEW')
```

```python
lst2 = []
for item in df['rate']:
    if(type(item) is not int and type(item) is not float):
        item = item.split('/')
        lst2.append(item)
    else:
        lst2.append(item)
```

```python
lst2
```

```python
lst3 = []
for item in lst2:
    if(type(item) is not int and type(item) is not np.nan and type(item) is not float):
        lst3.append(item[0])
    else:
        lst3.append(item)
```

```python
for i in range(len(lst3)):
    if(type(lst3[i]) is 'str'):
        lst3[i] = float(lst3[i])
    else:
        continue
lst3
```

```python
df['rate'] = lst3
```

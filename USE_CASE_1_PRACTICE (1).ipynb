```python
percent = (cabin_values*100)/total_values
```

```python
round(percent,3)
```

```python
percent_values_in_embarked = round(df2.Embarked.isnull().sum()*100/df2.shape[0], 3)
percent_values_in_embarked
```

```python
percent_values_in_age = round(df2.Age.isnull().sum()*100/df2.shape[0],3)
percent_values_in_age
```
